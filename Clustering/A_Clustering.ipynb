```python
import pandas as pd
import numpy as np
from google.colab import drive
import torch
from tqdm import tqdm
import re
import pickle
from ast import literal_eval
import json
import sys
from sklearn.cluster import AgglomerativeClustering
```

```python
path='/content/drive/MyDrive/종설/'
```

```python
net_E82=pd.read_csv(path+"network_E82.txt",header=None)
net_E91=pd.read_csv(path+"network_E91.txt",header=None)
```

```python
net_E82.columns=['user_idx']
net_E82['linked']=pd.Series()

for i in tqdm(range(len(net_E82))):
  net_E82.loc[i,'linked']=" ".join(net_E82.loc[i,'user_idx'].split(' ')[1:])
  net_E82.loc[i,'user_idx']=net_E82.loc[i,'user_idx'].split(' ')[0]
  
net_E82.head()
```

```python
net_E82['user_idx']=net_E82['user_idx'].astype('int')
```

```python
link_vector_E82=pd.DataFrame(index=net_E82['user_idx'],columns=net_E82['user_idx'])
link_vector_E82=link_vector_E82.fillna(0)
```

```python
for i in tqdm(link_vector_E82.index):
  a = list(map(int,net_E82.loc[net_E82['user_idx']==i,"linked"].str.split(' ').values[0]))
  for j in range(len(a)):
    link_vector_E82.loc[i,a[j]]=1
```

```python
link_vector_E82.to_csv(path+"link_vector_E82.txt",index=True,header=True)
```

```python
agg_g4 = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_predict_E82=agg_g4.fit_predict(link_vector_E82)
```

```python
pd.Series(cluster_predict_E82).value_counts()
```

```python
link_vector_E82['group']=cluster_predict_E82
```

```python
s_E82=pd.DataFrame(index=link_vector_E82.index,columns=range(1))
s_E82.columns=['grouped_user']

for i in tqdm(s_E82.index):
  li=[]
  for j in s_E82.index:
    if (link_vector_E82.loc[i,'group']==link_vector_E82.loc[j,'group']) & (i != j):
      li.append(j)
  s_E82.loc[i,'grouped_user']=li
  s_E82.loc[i,'grouped_user']=" ".join(list(map(str,s_E82.loc[i,'grouped_user'])))
```

```python
s_E82['user']=s_E82.index

for i in tqdm(s_E82.index):
  s_E82.loc[i,'grouped_user']=str(s_E82.loc[i,'user'])+" "+s_E82.loc[i,'grouped_user']
```

```python
s_E82=s_E82.drop('user',axis=1)
s_E82
```

```python
s_E82.to_csv(path+"S_E82_group4.txt",index=False,header=None)
```
