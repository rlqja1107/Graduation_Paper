```python
data = []
with open('..\\데이터\\Yelp\\item.json', 'r', encoding='UTF-8') as f:
    for line in f:
        data.append(json.loads(line))
item = pd.DataFrame(data)
```

```python
category_set = set()
for index, value in item.iterrows():
    if value['categories'] is not None:
        category_set.update(value['categories'].split(','))
```
